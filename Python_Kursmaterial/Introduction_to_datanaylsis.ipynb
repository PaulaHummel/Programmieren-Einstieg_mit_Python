```python
import matplotlib.pyplot as plt

umsatz = [100, 200, 300, 250, 400]
monate = ["Jan", "Feb", "Mär", "Apr", "Mai"]

plt.plot(monate, umsatz)
plt.title("Umsatzentwicklung")
plt.xlabel("Monat")
plt.ylabel("Umsatz in Tsd. €")
plt.show()
```

```python
# was ist ein dataframe: 
# Ein DataFrame ist in Python wie eine Excel-Tabelle
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    "Monat": monate,
    "Umsatz": umsatz
})

sns.barplot(data=df, x="Monat", y="Umsatz")
plt.title("Umsatz pro Monat")
plt.ylabel("Umsatz in Tsd. €")
plt.show()
```

```python
df = pd.DataFrame({
    "Monat": monate,
    "Umsatz": [100, 200, 300, 250, 400],
    "Gewinn": [20, 50, 70, 40, 90]
})

sns.scatterplot(data=df, x="Umsatz", y="Gewinn")
plt.title("Zusammenhang zwischen Umsatz und Gewinn")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Studiengang": ["BWL", "Jura", "Informatik", "Psychologie", "Sport"],
    "Kaffee pro Woche": [18, 22, 35, 25, 5]
})

farben = ["lightblue", "red", "black", "purple", "lightgreen"]

plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x="Studiengang",
    y="Kaffee pro Woche",
    hue="Studiengang",        
    palette=farben,
    legend=False               # damit keine doppelte Legende kommt
)
plt.title("Durchschnittlicher Kaffeekonsum pro Woche", fontsize=15)
plt.xlabel("Studiengang")
plt.ylabel("Tassen Kaffee")
plt.ylim(0, 40)
plt.tight_layout()
plt.show()
```
